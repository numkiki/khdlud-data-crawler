Đọc dữ liệu từ cellphones và thành các cột dữ liệu tương ứng

```python
import pandas as pd

df_cellphones = pd.read_csv('product_info_new.csv', encoding = 'utf-8')

df_temp = df_cellphones.pop('specs_info')
df_link = df_cellphones.pop('link')

fields_picked = ['Kích thước màn hình', 'Công nghệ màn hình', 'Camera sau', 'Camera trước', 'Chipset', 'Dung lượng RAM', 'Bộ nhớ trong',
          'Pin', 'Thẻ SIM', 'Hệ điều hành', 'Độ phân giải màn hình']

fields_cellphones = ['Tên điện thoại', 'Hãng', 'Giá tiền', 'Kích thước màn hình', 'Công nghệ màn hình', 'Camera sau', 'Camera trước', 'Chipset', 'Dung lượng RAM', 'Bộ nhớ trong',
          'Dung lượng pin', 'Số khe SIM', 'Hệ điều hành', 'Độ phân giải',  'Link', 'Nhà phân phối']

fields_new = ['Tên điện thoại', 'Hãng', 'Giá tiền', 'Nhà phân phối', 'Kích thước màn hình', 'Công nghệ màn hình', 'Camera sau', 'Camera trước', 'Chipset', 'Dung lượng RAM', 'Bộ nhớ trong',
              'Dung lượng pin', 'Số khe SIM', 'Hệ điều hành', 'Độ phân giải', 'Link']

#Tao dictionary moi
dt = {}
pos = len(df_cellphones.columns)
# Duyet qua tat ca field thong so muon lay
for field in fields_picked:
  dt[field] = []
  for x in df_temp:
    temp_dict = eval(x)
    # Kiem tra va khoi tao thong so bi thieu
    if field not in temp_dict.keys():
      temp_dict[field] = ''
    dt[field].append(temp_dict[field])
  df_cellphones.insert(loc=pos,column = field, value = dt[field])
  pos+=1

df_cellphones.insert(loc=pos,column = 'Link', value = df_link)
df_cellphones.insert(loc=pos+1,column = 'Nhà phân phối', value = 'Cellphones')
df_cellphones = df_cellphones.drop(columns=['original_price', 'rating', 'review_count'])
#Rename columns
df_cellphones.columns = fields_cellphones
#Reindex columns
df_cellphones = df_cellphones.reindex(columns = fields_new)
df_cellphones
```
